This notebook aims to investigate the WMI and MSBuild invokation incident.
It launches the Empire launcher.
First we load the dataset.

```kestrel
host = GET process
        FROM https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/lateral_movement/host/empire_msbuild_dcerpc_wmi_smb.zip
        WHERE command_line LIKE '%'
```

Since we know it utilizes WMI lets search for WMI

```kestrel
host_wmi = host WHERE command_line LIKE '%WMI%'
```

```kestrel
DISP host_wmi
```

Lets see if it have any child process. Turns out they have 1 entity with network traffic. Lets display it first.

```kestrel
wmi_child = FIND process CREATED BY host_wmi
```

```kestrel
DISP wmi_child
```

The command line executes MSBuild. Lets see if the child process, which should be msbuild have any network traffic.
Another point to note, that it loads an xml to do its thing. Possibly contains instructions and such.

```kestrel
msbuild = FIND process CREATED BY wmi_child
```

```kestrel
DISP msbuild
```

we found the msbuild, and from here we know there's some traffic happening here. lets see if we can extract the network activity.

```kestrel
net_msbuild = FIND network-traffic CREATED BY msbuild
```

turns out there's no network traffic associated with both process, which is weird. probably both process is triggered remotely from another process, so lets see WMI's parents

```kestrel
wmi_parent = FIND process CREATED host_wmi
```

Here we found another DCOM launch, which indicates that the svchost is run remotely.

```kestrel
DISP wmi_parent
```

Still no luck. Maybe the traffic is somewhere from the host process?

```kestrel
netms = FIND network-traffic CREATED BY wmi_parent
```

Gottem!

```kestrel
nethost = FIND network-traffic CREATED BY host
```

```kestrel
net_proc = FIND process CREATED nethost
```

Turns out its the powershell!

```kestrel
DISP net_proc
```

```kestrel
net_child = FIND process CREATED net_proc
```

And the child process chekcs out! So in a sense, we can trace that the malicious actor connects remotely with DCOM, loads XML file with powershell that contains instructions to invoke MSBuild and WMI processes.

```kestrel
DISP net_child
```
