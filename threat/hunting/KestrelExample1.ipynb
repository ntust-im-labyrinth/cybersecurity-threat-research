# Example 1: Empire Launcher

In this example there is an VBS Script used as a launcher for initial access. The adversary uses obfiscation to hide the script and make it difficult to analyze.

Source: https://securitydatasets.com/notebooks/atomic/windows/execution/SDWIN-190518182022.html?highlight=launcher%20vbs

Dataset: https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip

### Get Processes

First we get all processes from the dataset.

```kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip where command_line LIKE '%'
```

### Filter and search for "WMICS"
None processes are found.

```kestrel
sprocs = procs where command_line like '%WMICS%'
```

```kestrel
DISP sprocs
```

### Filter and search for "vbs"

Since we know, that an output file with the name "launcher.vbs" is generated, we search for that.

```kestrel
sprocs = procs where command_line like '%vbs%'
```

We found one suspect process and display it.

```kestrel
DISP sprocs
```

The launcher script can ben found on the Desktop (C:\Users\pgustavo\Desktop\launcher.vbs).

### Filter and search for Process name
We know from the dataset description, that the process is running in the powershell.

This opens up another approach to find the launcher.vbs script. To exclude all the uninteresting processes we can only search for the ones with a "created" stamp.

```kestrel
#susprocess = GET process FROM procs WHERE [process:pid = '2316 ']
#DISP susprocess ATTR name, pid
susprocess = GET process FROM procs WHERE [process:name = 'powershell.exe' AND process:created != 'None']
DISP susprocess ATTR name, pid, parent_ref.command_line
```

```kestrel
DISP susprocess
```

We can also search for "wscript.exe".

```kestrel
susprocess = GET process FROM procs WHERE [process:name = 'wscript.exe']
DISP susprocess
```

### Another approach: Checking the logs
Found nothing here. Deobfiscation with logs4shell did not work.

```kestrel
logs = GET artifact FROM https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip
WHERE [artifact:payload_bin MATCHES '.*']
```

```kestrel
susp = GET artifact FROM logs WHERE [artifact:exploit LIKE '%']
DISP susp ATTR id, first_observed, exploit, original
```
