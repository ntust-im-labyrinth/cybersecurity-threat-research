<h1>Loading All Processes</h1>

```kestrel
procs = GET process 
    from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip 
    where command_line 
    LIKE '%'
```

<h1>Filter them</h1>

```kestrel
sprocs = procs WHERE command_line LIKE '%WMIC%'
```

<h1>Filter them better and display results</h1>

```kestrel
sprocs = procs WHERE command_line LIKE '%vbs%'
DISP sprocs
```

<h1>Go deeper into suspicious process</h1>
<p>- find the user, connected processes, and network-traffic</p>

```kestrel
DISP sprocs ATTR creator_user_ref.user_id 
```

```kestrel
sprocess = GET process 
                FROM procs
                WHERE [process:creator_user_ref.user_id = "THESHIRE\pgustavo"]

DISP sprocess
```

```kestrel
snt = FIND network-traffic CREATED BY sprocess
DISP snt
```

```kestrel
snt_out = GET process 
        FROM procs 
        WHERE [network-traffic:src_ref.value = '172.18.39.5']
```

```kestrel
snt_in = GET process 
        FROM procs 
        WHERE [network-traffic:dst_ref.value = '172.18.39.5']
```

```kestrel
DISP snt_in
```
