# Empire Remote WMIC Add User

## Description: This dataset represents an adversary remotely executing code via WMI to ad a backdoor user on the target system. This dataset focuses on the use of the WMI Win32_Process class and method Create to execute code remotely.

https://securitydatasets.com/notebooks/atomic/windows/execution/SDWIN-200914080546.html



```kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/lateral_movement/host/empire_wmic_add_user_backdoor.zip
where command_line LIKE '%'
```

## Find processes that could create a backdoor and end with "exe"

```kestrel
filtered_procs = GET process FROM procs WHERE [process:command_line LIKE '%exe%']
```

## Get an overview of all processes with "exe" as extension

```kestrel
DISP filtered_procs
```

## See which parent process has executed a command with "backdoor"

```kestrel
backdoor = GET process FROM filtered_procs WHERE parent_ref.command_line LIKE '%backdoor%'
```

## Get a better overview of the two entities

```kestrel
DISP backdoor
INFO backdoor
```

## It does not appear to be the case that network traffic was generated in relation to the backdoor process. 

```kestrel
nt = FIND network-traffic CREATED BY backdoor
```

```kestrel
parent_procs = FIND process CREATED backdoor
```

## Process 3692 shows that a user named backdoor has been added

```kestrel
DISP backdoor ATTR pid, name, id, command_line, created
```
