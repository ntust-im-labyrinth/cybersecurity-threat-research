# Example 2: Add Backdoor user

In this example a hacker is creating a backdoor user. For that WMI is used to remotley executing code in the system.

In the dataset the WMI Win32_Process and the method "create" is used.


Source: https://securitydatasets.com/notebooks/atomic/windows/execution/SDWIN-200914080546.html

Dataset: https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/lateral_movement/host/empire_wmic_add_user_backdoor.zip

### Get Processes

First we get all processes from the dataset.

```kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/lateral_movement/host/empire_wmic_add_user_backdoor.zip where command_line LIKE '%'
```

### Filter and search for "WMIC"
We search for WMIC. WMIC is a tool that allows command-line access to the Windows Management Instrumentation (WMI). With this tool, it is possible to query the operating system for detailed information about installed hardware and Windows settings, run management tasks, and even execute other programs or commands. In this case it was used for the last case, executing commands.

```kestrel
sprocs = procs where name like '%WMIC%'
DISP sprocs
```

### Filter and search for the command
An other approach is to search in the command line for the command "net user /add" directly. This method could be used in threat hunting to monitor unwanted system changes.

```kestrel
sprocs = procs where command_line like '%net user%'
DISP sprocs
```

### Check the pid mentioned in the description
We can also just look for the mentioned process id, to see what the process looks like to see, what hunting patterns we can apply. (Good first step when a data set is avilable, not useful in general threat hunting.)

```kestrel
susprocess = GET process FROM procs WHERE [process:pid = '7768']
DISP susprocess
```
