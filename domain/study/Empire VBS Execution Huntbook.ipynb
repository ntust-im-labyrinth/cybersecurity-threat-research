## Explore Datasets

### Download & Decompress Dataset

```kestrel
import requests
from zipfile import ZipFile
from io import BytesIO

url = "https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip"
zipFileRequest = requests.get(url)
zipFile = ZipFile(BytesIO(zipFileRequest.content))
datasetJSONPath = zipFile.extract(zipFile.namelist()[0])
```

### Read JSON File

```kestrel
from pandas.io import json

df = json.read_json(path_or_buf=datasetJSONPath, lines=True)
```

### Access Security Events

```kestrel
df.groupby(['Channel']).size().sort_values(ascending=False)
```

# Adversary View

```kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip where command_line LIKE '%'
```

## Find relate process extension with ".exe"

```kestrel
sprocs = procs where command_line LIKE '%exe%'
```

### This means 7 differents processes are runing in this dataset. So now let's display all of them

```kestrel
DISP sprocs
```

### Well, we agree, there is a size indicator ;). So, let's take a closer look at what powershell.exe is doing. In other words, let's see at what this generates (childs)

```kestrel
task1 = FIND process CREATED BY procs
powershell = GET process FROM task1 WHERE [process:name = 'powershell.exe']
powershell_child = FIND process CREATED by powershell
DISP powershell_child ATTR name, command_line
```

### And now let's found what generated this .exe (parent)

```kestrel
powershell_parent = FIND process CREATED powershell
DISP powershell_parent ATTR name, command_line
```

### Now let's look at the files that powershell.exe processes

```kestrel
powershell_docs = FIND file LINKED powershell
INFO powershell_docs
```

```kestrel
DISP powershell_docs ATTR name, parent_directory_ref.path
```

### Then powershell's childs

```kestrel
powershell_child_docs = FIND file LINKED powershell_child
INFO powershell_child_docs
```

```kestrel
DISP powershell_child_docs ATTR name, parent_directory_ref.path
```

### And finally powershell's parent

```kestrel
powershell_parent_docs = FIND file LINKED powershell_parent
INFO powershell_parent_docs
```

```kestrel
DISP powershell_parent_docs ATTR name, parent_directory_ref.path
```
