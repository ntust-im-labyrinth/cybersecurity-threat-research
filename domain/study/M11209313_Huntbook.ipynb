```kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/execution/host/empire_launcher_vbs.zip where command_line LIKE '%'
```

```kestrel
sprocs = procs where command_line LIKE '%cmd%'
```

```kestrel
sprocs = procs where command_line LIKE '%bat%'
```

```kestrel
sprocs = procs where command_line LIKE '%powershell%'
```

```kestrel
DISP sprocs
```

```kestrel
tasks = FIND process CREATED BY procs
powershell = GET process FROM tasks WHERE [process:name = 'powershell.exe']
powershell_parent = FIND process CREATED powershell
DISP powershell_parent ATTR name, command_line
```

```kestrel
amount = FIND network-traffic CREATED BY powershell
DISP amount ATTR hash.md5
```

```kestrel
amount1 = FIND network-traffic CREATED BY powershell
DISP amount1 ATTR direction
```
