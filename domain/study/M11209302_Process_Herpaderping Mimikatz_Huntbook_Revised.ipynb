# Process Herpaderping Mimikatz

This dataset represents the execution of a Process Herpaderping to obscure the intentions of a process by modifying the content on disk after the image has been mapped.

# Tactics

Process Injection [TA0004](<https://attack.mitre.org/tactics/TA0004/>), Defense Evasion [TA0005](<https://attack.mitre.org/tactics/TA0005/>)

# Techniques

Process Injection [T1055](<https://attack.mitre.org/techniques/T1055/>)

````kestrel
procs = GET process from https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/defense_evasion/host/cmd_process_herpaderping_snippingtool.zip where command_line LIKE '%'
````

# Find relate process extension with "exe"

````kestrel
sprocs = procs where command_line LIKE '%exe%'
````

# There are 4 process are running in this dataset. And we found ProcessHerpadering.exe is excuting by mimikatz.exe from command line

````kestrel
DISP sprocs
````

# Find child processes created by ProcessHerpaderping.exe to see what are these process doing.

````kestrel
tasks = FIND process CREATED BY procs
ProcessHerpaderping = GET process FROM tasks WHERE [process:name = 'ProcessHerpaderping.exe']
ProcessHerpaderping_child = FIND process CREATED by ProcessHerpaderping
DISP ProcessHerpaderping_child ATTR name, command_line
````

# Investigate whether this suspious process had any network traffic

````kestrel
networktraffic = FIND network-traffic CREATED BY ProcessHerpaderping
DISP networktraffic ATTR dst_ref.value, dst_port
````

# Take a deeper look to child process of ProcessHerpadering's child process

````kestrel
ProcessHerpaderping_child_child = FIND process CREATED BY ProcessHerpaderping_child
DISP ProcessHerpaderping_child_child ATTR name, command_line
````

# This step is to see what files are read/written/excuted by ProcessHerpadering_child_child.

> FIND : wardog.exe

````kestrel
ProcessHerpaderping_file = FIND file LINKED ProcessHerpaderping_child_child
DISP ProcessHerpaderping_file ATTR name, parent_directory_ref.path
````

# This step is to see what are the hostname and events that generated by ProcessHerpadering_child_child.

> FIND : wardog.exe

````kestrel
hosts = FIND x-oca-asset LINKED ProcessHerpaderping_child_child
DISP hosts
````

````kestrel
events = FIND x-oca-event LINKED ProcessHerpaderping_child_child
DISP events
````
