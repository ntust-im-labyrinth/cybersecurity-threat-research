### Dataset Description
This dataset represents a threat actor using Seatbelt profiling an endpoint. This specifically uses the -group=user command.

Looking for command_line LIKE '%Seatbelt%':

```kestrel
procs = get process from 
https://raw.githubusercontent.com/OTRF/Security-Datasets/master/datasets/atomic/windows/discovery/host/cmd_seatbelt_group_user.zip 
where command_line LIKE '%Seatbelt%' 
```

Find parent process:

```kestrel
assets = find x-oca-asset linked procs
disp assets
events = find x-oca-event linked procs
disp events
```
